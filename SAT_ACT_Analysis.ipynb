#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
# Function to calculate standard deviation
import math
def my_stddev(my_list):
    my_sum = sum(my_list)
    
    total_no = len(my_list)
    my_avg = my_sum / total_no
    
    sum_avg = 0.0
    for i in my_list:
        sum_avg =sum_avg + math.pow((i - my_avg), 2)
        
    my_variance = sum_avg / total_no
    
    std = math.sqrt(my_variance)
    return std
````

- Use a **dictionary comprehension** to apply your standard deviation function to each numeric column in the dataframe.  **No loops**  
- Assign the output to variable `sd` as a dictionary where: 
    - Each column name is now a key 
    - That standard deviation of the column is the value 
     
*Example Output :* `{'ACT_Math': 120, 'ACT_Reading': 120, ...}`

````python
# Function to plot histogram
def subplot_histograms(dataframe, list_of_columns, list_of_titles, list_of_xlabels):
    nrows = int(np.ceil(len(list_of_columns)/2)) # Setting 2 colums per row
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=(15, 12)) # Setting figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        ax[i].hist(dataframe[column])# Plotting the graph
        ax[i].set_title(list_of_titles[i], size = 16)# Setting title
        ax[i].set_xlabel(list_of_xlabels[i], size = 12)# Setting x label
````

#### Plot and interpret boxplots

For each numeric variable in the dataframe create a boxplot using Seaborn. Boxplots demonstrate central tendency and spread in variables. In a certain sense, these are somewhat redundant with histograms, but you may be better able to identify clear outliers or differences in IQR, etc.

Multiple values can be plotted to a single boxplot as long as they are of the same relative scale (meaning they have similar min/max values).

Each boxplot should:
- Only include variables of a similar scale
- Have clear labels for each variable
- Have appropriate titles and labels

````python
# Function to plot boxplots
def subplot_boxplot(dataframe, list_of_titles, list_of_xlabels, list_of_ylabels, list_of_xticklabels):
    ncols = len(dataframe) #Setting columns
    nrows = int(np.ceil(len(dataframe)/2)) # Setting 2 colums per row
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20, 10)) # Setting figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    sns.set_palette('pastel') #Setting color
    for i in range(ncols): # Gives us an index value to get into all our lists
        ax[i] = sns.boxplot(x="variable", y="value", data=pd.melt(dataframe[i]), ax = ax[i]) # Plotting the graph
        ax[i].set_title(list_of_titles[i], fontsize = 20) # Set titles
        ax[i].set_xlabel(list_of_xlabels[i], fontsize = 16) # Set x lables
        ax[i].set_ylabel(list_of_ylabels[i], fontsize = 16) # Set y labels
        ax[i].tick_params(axis="x", labelsize = 14) # Set x ticks
        ax[i].tick_params(axis="y", labelsize = 14) # Set y ticks
        ax[i].set_xticklabels(list_of_xticklabels) # Set x tick labels

# https://www.codecademy.com/articles/seaborn-design-ii
# https://www.geeksforgeeks.org/python-pandas-melt/
# https://stackoverflow.com/questions/49554139/boxplot-of-multiple-columns-of-a-pandas-dataframe-on-the-same-figure-seaborn
````
